```python
def drop_z(x, y, R=3):
    r = np.sqrt(x**2 + y**2)
    r_norm = r / R
    r_norm = np.clip(r_norm, 0, 1)
    v_angle = (np.pi / 2) * (1 - r_norm)
    z = np.where(v_angle >= 0,
                 R * np.sin(v_angle) + R * (v_angle / (np.pi / 2))**4,
                 R * np.sin(v_angle))
    return z

z_coords = drop_z(coords[:, 0], coords[:, 1])
```
